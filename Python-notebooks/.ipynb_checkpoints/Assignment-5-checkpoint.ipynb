Question 1: Available subscriber types Subscriber = annual or 30-day member; Customer = 24-hour or 3-day member We could
recommend another subscriber type for RUSH HOUR CUSTOMER

Answer: Total no of trips and No of trips in the rush hour are below

```python
! bq query --use_legacy_sql=false 'SELECT COUNT(*) as TOTAL_NO_OF_TRIPS FROM \
`bigquery-public-data.san_francisco.bikeshare_trips`'
```

```python
! bq query --use_legacy_sql=false 'SELECT COUNT(*) as RUSH_HOUR_TRIPS \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
WHERE \
  EXTRACT(DATE from start_date)= EXTRACT(DATE from end_date) and \
  ((EXTRACT(HOUR from start_date) between 7 and 9 and \
  EXTRACT(HOUR from end_date) between 7 and 9) or (EXTRACT(HOUR from start_date) between 16 and 19 and \
  EXTRACT(HOUR from end_date) between 16 and 19))'
```

Recommendation 1 : Ford GoBike only has 2 subscriber types, Subscriber = annual or 30-day member; Customer = 24-hour or 3-day member, 60 % of the total Trips are in the RUSH hour, A new subscriber type called RUSH HOUR CUSTOMER can be proposed which could largely the increase customer base.

Question 2: What are the most popular start and end stations in rush hour/commuter trips

Answer: The rush hours are morning 7 to 9 AM and evening 4 to 7 PM. The top 10 most popular stations are below

```python
! bq query --use_legacy_sql=false --format=csv ' SELECT COUNT(*) as NO_OF_TRIPS, \
      START_STATION_NAME, \
      END_STATION_NAME \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
WHERE \
  EXTRACT(DATE from start_date)= EXTRACT(DATE from end_date) and \
  ((EXTRACT(HOUR from start_date) between 7 and 9 and  \
  EXTRACT(HOUR from end_date) between 7 and 9) or (EXTRACT(HOUR from start_date) between 16 and 19 and \
  EXTRACT(HOUR from end_date) between 16 and 19)) \
group by START_STATION_NAME, END_STATION_NAME \
ORDER by NO_OF_TRIPS DESC limit 10' > rushhourtrips.csv
```

```python
rush_hour_trips = pd.read_csv('rushhourtrips.csv')
rush_hour_trips
```

Recommendation 2 : As most riders have the above start station and end station combination there could be cases where not enough bikes are available during rush hours, a recommendation to increase the number of Bikes(also Docks) helps to increase customer base and helps with customer satisfaction

Question 3: Propose new subscriber types in the most popular Stations where subscriber type is not Subscriber based on total no of rush hour trips

Answer: Most of the Rush hour trips have subscriber type = Subscriber, who is a annual or 30-day member, the top 10 entries as shown below indicates that a good amount of customers are with subscriber type = Customer in the rush hour, this would enable us to recommend a new subscriber type.

```python
! bq query --use_legacy_sql=false --format=csv 'SELECT COUNT(*) as NO_OF_TRIPS, \
START_STATION_NAME,END_STATION_NAME, SUBSCRIBER_TYPE FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
WHERE subscriber_type = "Customer" and \
(EXTRACT(DATE from start_date)= EXTRACT(DATE from end_date) and ((EXTRACT(HOUR from start_date) between 7 and 9 \
and EXTRACT(HOUR from end_date) between 7 and 9) or (EXTRACT(HOUR from start_date) between 16 and 19 \
and EXTRACT(HOUR from end_date) between 16 and 19))) \
group by START_STATION_NAME, END_STATION_NAME,SUBSCRIBER_TYPE \
ORDER by NO_OF_TRIPS DESC limit 10' > rushhourtripsst.csv
```

```python
rush_hour_trips_subscriberType = pd.read_csv('rushhourtrips.csv')
rush_hour_trips_subscriberType
```

Recommendation 3 : A new subscriber types can be introduced targetting these customers which are Half Day pass and RUSH HOUR CUSTOMER(Recommendation 1).

Question 4: We could provide discounted rates in most popular station landmarks.

Answer: First query talks about landmark associated to all trips, second query talks about total of stations in each landmark.

```python
! bq query --use_legacy_sql=false --format=csv 'SELECT trip_id, START_STATION_ID, START_STATION_NAME,\
(SELECT landmark FROM `bigquery-public-data.san_francisco.bikeshare_stations` as bikeshare_stations \
where bikeshare_stations.STATION_ID = bikeshare_trips.START_STATION_ID) as START_STATION_LANDMARK,\
END_STATION_ID,  END_STATION_NAME, \
(SELECT landmark FROM `bigquery-public-data.san_francisco.bikeshare_stations` as bikeshare_stations \
WHERE bikeshare_stations.STATION_ID = bikeshare_trips.END_STATION_ID) as END_STATION_LANDMARK \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` as bikeshare_trips limit 50' > landmarks.csv
```

```python
landmarks = pd.read_csv('landmarks.csv')
landmarks
```

```python
! bq query --use_legacy_sql=false --format=csv 'SELECT count(*) AS NO_OF_STATIONS ,LANDMARK \
FROM `bigquery-public-data.san_francisco.bikeshare_stations`\
GROUP BY LANDMARK ORDER BY NO_OF_STATIONS DESC' > landmarkcount.csv
```

```python
landmark_count = pd.read_csv('landmarkcount.csv')
landmark_count
```

Recommendation 4 : Discounted offers or day passes in most popular stations would increase customer base
